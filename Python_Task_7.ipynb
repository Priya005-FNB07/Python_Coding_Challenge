#While Loop,Else,Break,Continue:

##Shopping Cart System:

```python
total_amount=0
item_count=0
print("enter the product price:")
while True:
  price=float(input())
  if price==0:
      break
  if price<0:
    print("no negative prices are allowed")
    continue
  total_amount+=price
  item_count+=1
if total_amount>=1000:
   discount = 0.20
elif total_amount>=500:
   discount =0.10
elif total_amount>=200:
   discount=0.05
else:
   discount=0
total_discount=total_amount*discount
final_amount=total_amount-total_discount
print('total_item:',item_count)
print('sub_total:',total_amount)
print('discount_amount:',total_discount)
print('final_total:',final_amount)
```
