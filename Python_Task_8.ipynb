#Range,for loop,Enumerate,List Comprehension:

```python
exam_scores=[57,78,67,91,53,40,87,77,86,51]
print(exam_scores)
```

```python
for index,exam_scores in enumerate(exam_scores):
    print(index,':',exam_scores)
```

```python
exam_scores = [57,78,67,91,53,40,87,77,86,51]
high_scores = [score for score in exam_scores if score >= 75]
print(high_scores)
```

```python
average=sum(exam_scores)/len(exam_scores)
print(average)
```
